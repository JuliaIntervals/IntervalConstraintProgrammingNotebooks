```julia
using ValidatedNumerics, IntervalConstraintProgramming
```

```julia
c, d, e, h = 1, 1, 1, 1

@function f(x,y) = x*y
@function g(x) = $c*x^3 + $d*x^2 + $e*x + $h   # errors if defined as g(x,y) and doesn't contain y

C = @constraint f(x,y) - g(x) == 0
X = IntervalBox(-4..4, -4..4)
```

```julia
@time paving = pave(C, X, 0.02)
```

```julia
using Plots
gr()
```

```julia
plot(paving.boundary, aspect_ratio=:equal, c="gray")
```
